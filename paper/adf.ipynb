```python
from chebgreen.greenlearning.utils import approximateDistanceFunction
import numpy as np
import torch
import matplotlib.pyplot as plt

device = torch.device('cpu')
```

```python
xx = np.linspace(-4,2,2000)
yy = np.linspace(-4,2,2000)
x, y = np.meshgrid(xx,yy)
x = torch.from_numpy(x).to(device)
y = torch.from_numpy(y).to(device)
ADF = approximateDistanceFunction(x, y, np.array([-4, 2, -4, 2]), device)
```

```python
x = x.cpu().numpy()
y = y.cpu().numpy()
ADF = ADF.cpu().numpy()
```

```python
np.max(ADF)
```

```python
fig = plt.figure(figsize = (13,10), frameon=False)
plt.axis('off')
vmin = 0
vmax = 1.4
plt.gca().set_aspect('equal', adjustable='box')
levels = np.linspace(vmin, vmax, 50, endpoint = True)
plt.contourf(x,y,ADF, levels = levels, cmap = 'jet', vmin = vmin, vmax = vmax)
# ticks = np.linspace(vmin, vmax, 5, endpoint=True)
# cbar = plt.colorbar(ticks = ticks, fraction = 0.046, pad = 0.04)

fig.savefig('ADF.png', dpi = fig.dpi, bbox_inches='tight', pad_inches=0)
# savePath = f"plots/{example}"
# Path(savePath).mkdir(parents=True, exist_ok=True)
# fig.savefig(f'{savePath}/{example}-{label}.png', dpi = fig.dpi, bbox_inches='tight', pad_inches=0)
```
